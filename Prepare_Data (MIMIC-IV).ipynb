```python
radiology_df = pd.read_csv("radiology.csv")
```

```python
radiology_df.head()
```

```python
ids = df.hadm_id.tolist()
```

```python
rad_df = radiology_df[radiology_df.hadm_id.isin(ids)]

df_min = (
    rad_df[rad_df['text'].str.startswith('INDICATION', na=False)]
    .loc[lambda x: x.groupby('hadm_id')['note_seq'].idxmin()]
    .reset_index(drop=True)
)
```

```python
df_min
```

```python
df_min = pd.merge(df_min, df[['hadm_id', 'sepsis']], on='hadm_id') 
```

```python
df_min.sepsis.value_counts()
```

```python
df_min.head()
```

```python
df_balanced = df_min.copy()
```

### Balance 

```python
# Separate classes
df_majority = df_min[df_min['sepsis'] == 0]
df_minority = df_min[df_min['sepsis'] == 1]

# Downsample majority
df_majority_down = df_majority.sample(
    n=len(df_minority),
    random_state=42
)

# Combine and shuffle
df_balanced = (
    pd.concat([df_majority_down, df_minority])
    .sample(frac=1, random_state=42)
    .reset_index(drop=True)
)
```

```python
df_balanced.to_csv("radiology_notes.csv", index=False)
```

### Tabular Data 

```python
tabular_df = df[df.hadm_id.isin(df_balanced.hadm_id.tolist())]
```

```python
tabular_df = pd.merge(tabular_df, df_balanced[['hadm_id', 'text']], on='hadm_id')
```

```python
tabular_df.sepsis.value_counts()
```

```python
from sklearn.model_selection import train_test_split

# --------------------------
# Step 1: Train 60%, Temp 40%
# --------------------------
train_df, temp_df = train_test_split(
    tabular_df,
    test_size=0.3,
    stratify=tabular_df["sepsis"],
    random_state=42
)

# --------------------------
# Step 2: Split temp → DSEL 20%, Test 20%
# --------------------------
dsel_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["sepsis"],
    random_state=42
)
```

```python
train_df.to_csv("sepsis_data/train.csv", index=False) 
dsel_df.to_csv("sepsis_data/dsel.csv", index=False) 
test_df.to_csv("sepsis_data/test.csv", index=False) 
```
